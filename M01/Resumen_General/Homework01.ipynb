```python
def NumeroBinario(numero):
    '''
    Esta función recibe como parámetro un número entero mayor ó igual a cero y lo devuelve en su 
    representación binaria. Debe recibir y devolver un valor de tipo entero.
    En caso de que el parámetro no sea de tipo entero y mayor a -1 retorna nulo.
    '''
    if type(numero) != int:
        return None
     
    if numero < 0:
        return None

    elif numero == 0:
        return 0

    lista = []                   # Almacenamiento de los restos
    div_resto = 0

    while (numero > 0):          # Mientras que el numero sea mayor que 0, vamos a hacer la division sucesiva
        div_resto = numero % 2   # Sacamos el resto de dividir por dos
        numero = numero // 2     # Se divide sucesivamente entre dos
        lista.append(div_resto)  # Guardamos el resto conseguido

    numero_binario = ''           # Esto vamos a retornar
    for elemento in lista:
        numero_binario = str(elemento) +  numero_binario 
    
    # lista = lista[::-1]

    return int(numero_binario)
```

```python
def NumeroBinarioFraccionario(decimal):
    if (decimal >= 1) or (decimal < 0):
        return None
    elif decimal == 0:
        return 0
    
    lim_periodico = 24               # Número máximo de decimales
    lista = []

    while decimal > 0 and lim_periodico > 0:
        lista.append(int(decimal * 2))
        decimal = decimal * 2 - (int(decimal * 2))
        lim_periodico -= 1

    numero_binario = '0.'
    for elemento in lista:
        numero_binario = numero_binario + str(elemento)

    return numero_binario
```
