```python
import pandas as pd

data = pd.read_csv(r"3 Juegos en steam.csv")
```

```python
# Assuming 'genres' contains the semicolon-separated genres
data['genres'] = data['genres'].str.split(';')

# Create a new DataFrame by exploding the genres
exploded_data = data.explode('genres')

# Count the occurrences of each genre
genre_counts = exploded_data['genres'].value_counts()

# Display the result
print(genre_counts)
```

```python
# Assuming 'categories' contains the comma-separated categories
data['categories'] = data['categories'].str.split(';')

# Create a new DataFrame by exploding the categories
exploded_data = data.explode('categories')

# Count the occurrences of each category
category_counts = exploded_data['categories'].value_counts()

# Display the result
print(category_counts)
```

```python
# Assuming 'genres' and 'categories' contain the appropriate data
data['genres2'] = data['genres'].str.split(';')
data['categories2'] = data['categories'].str.split(';')

# Explode the genres and categories
exploded_data = data.explode('genres2').explode('categories2')

# Group by 'genres' and count unique 'categories' within each group
genre_category_counts = exploded_data.groupby(['genres2', 'categories2'])['categories2'].count()

# Display the result
print(genre_category_counts)
```

```python
# Set pandas display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Now, when you print the result, it should show all rows
print(genre_category_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"3 Juegos en steam.csv")

# Assuming 'genres' contains the semicolon-separated genres
data['genres'] = data['genres'].str.split(';')

# Create a new DataFrame by exploding the genres
exploded_data = data.explode('genres')

# Count the occurrences of each genre
genre_counts = exploded_data['genres'].value_counts()

# Calculate the total count
total_count = genre_counts.sum()

# Filter categories with percentage less than 5%
threshold = 0.05  # 5%
filtered_genre_counts = genre_counts[genre_counts / total_count >= threshold]

# Group categories with percentage less than 5% as "Other"
filtered_genre_counts['Other'] = genre_counts[genre_counts / total_count < threshold].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(filtered_genre_counts, labels=filtered_genre_counts.index, autopct='%1.1f%%', startangle=140, labeldistance=1.2)
plt.title('Genre Distribution')

# Show the pie chart
plt.show()
```
