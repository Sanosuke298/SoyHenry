```python
import pandas as pd
import numpy as np
import nltk #Procesamiento lenguaje corporal
import os
nltk.download('punkt')

import nltk.corpus

text = """Ingrese letra de canción acá"""

from nltk.tokenize import word_tokenize

token = word_tokenize(text)
token
```

# Eliminando Puntuacion

```python
import string
new_words = []
for word in token:
    for letter in word:
        if letter in string.punctuation:
            word = word.replace(letter,"")
    new_words.append(word)

print(new_words)
```

# Eliminando espacios vacios

```python
new_list = []
for vacio in new_words:
  if vacio != "":
    new_list.append(vacio)
print(new_list)
```

```python
minusculas=[]
for i in new_list:
    i = i.lower()
    minusculas.append(i)
print(minusculas)
```

# Eliminando StopWords

```python
from nltk import word_tokenize
from nltk.corpus import stopwords

nltk.download("stopwords")

a = set(stopwords.words('english'))
#print(a) #Muestra las palabras que son vacias para el ingles

stopwords = [x for x in minusculas if x not in a]
print(stopwords)
```

# Frecuencia de palabras

```python
#Frecuencia de valores en el texto
from nltk.probability import FreqDist
fdist = FreqDist(stopwords)
fdist
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

text = " ".join(stopwords)

wordcloud = WordCloud(width = 1000, height = 1000, background_color = "white", stopwords = a , min_font_size=10).generate(text)

plt.figure(figsize= (8,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
